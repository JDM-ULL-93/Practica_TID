# Preprocesamiento de datos

Nuestros datos no tienen valores nulos por lo tanto no hay ni que analizar ni imputarle valores a ningún campo nulo . Pero si presenta un gran número de valores outliers o valores atipicos. Vamos a proceder a remover la mayoría


## Analisís valores atipicos .

Antes de seguir, vamos a dividir nuestro datasheet por cada clase presente, para así disminuir el error en la detección de valores outliers.

```r
data0 <- data[which(data[,65] == 0), 1:65 ]
data1 <- data[which(data[,65] == 1), 1:65 ]
data2 <- data[which(data[,65] == 2), 1:65 ]
data3 <- data[which(data[,65] == 3), 1:65 ]
```

Usaremos el algoritmo **LOF** de la libreria **DDoutlier** para calcular un valor entre 0 e inf, de forma que, cuanto mas cercano al 0 menos atipico (por lo tanto más atipico cuanto mayor)

Más documentación puede encontrarse [aqui](https://cran.r-project.org/web/packages/DDoutlier/DDoutlier.pdf)

```r
if (!requireNamespace("DDoutlier"))
  install.packages("DDoutlier")
library(DDoutlier)
```

```r
outlier_score0 <- LOF(data0, k= 40)
outlier_score1 <- LOF(data1, k= 40)
outlier_score2 <- LOF(data2, k= 40)
outlier_score3 <- LOF(data3, k= 40)
```

El algoritmo previo nos genera una puntuación para cada fila de nuestro conjunto de datos, de forma que:
- Una fila cercana a 1 representa una fila dentro de la densidad normal del conjunto de datos global
- Menor a 1 presenta una densidad mayor respecto al promedio
- Cuanto mayor a 1, una densidad mucho menor al promedio, es decir, valores alejados de un centro de gravedad y, por lo tanto, outliers. 

Con el siguiente codigo vamos a mostrar un histograma que muestra la frecuencia de esta puntuación para todo nuestro conjunto de datos:

```r
data_outlier0 <- cbind(data0, outlier_score0)
data_outlier1 <- cbind(data1, outlier_score1)
data_outlier2 <- cbind(data2, outlier_score2)
data_outlier3 <- cbind(data3, outlier_score3)

hist(data_outlier0[,66])
hist(data_outlier1[,66])
hist(data_outlier2[,66])
hist(data_outlier3[,66])
```

Vemos mas o menos que los valores atipicos empiezan:
- Para clase 0 , en score >= ~2.0
- Para clase 1, en score >= ~1.4
- Para clase 2, en score >= ~1.6
- Para clase 3, en score >= ~1.8

Según el criterio anterior, empezamos a seleccionar, para cada datasheet, las filas con un score inferior y volvemos a combinar todos los datasheet en uno solo

```r
dataCleaned0 <- data_outlier0[which(data_outlier0[,66] <= 1.7), 1:65 ]
dim(dataCleaned0)
dataCleaned1 <- data_outlier1[which(data_outlier1[,66] <= 1.3), 1:65 ]
dim(dataCleaned1)
dataCleaned2 <- data_outlier2[which(data_outlier2[,66] <= 1.5), 1:65 ]
dim(dataCleaned2)
dataCleaned3 <- data_outlier3[which(data_outlier3[,66] <= 1.4), 1:65 ]
dim(dataCleaned3)

dataCleaned <- rbind(dataCleaned0,dataCleaned1,dataCleaned2,dataCleaned3)

dim(dataCleaned)
#outliers <- 0
par(mfrow=c(1,4))
for(i in 1:65) {
  lista <- boxplot(dataCleaned[,i], main=names(dataCleaned)[i])
  #outliers <- outliers + length(lista$out)
}
#print("Numero de outliers segun boxPlot =")
#print(outliers/64)
```

## Analisis de correlación

Una vez hemos limpiado parte de nuestros valores outliers, procedemos a realizar el analisís de componentes principales mediante el algoritmo RandomForest.

Lo primero, es visualizar una matriz de graficas de correlación. Como siempre, lo primero es cargar las librerias necesarias

```r
if ( !requireNamespace("corrplot") ) {
  install.packages("corrplot")
}

library(corrplot)
```

Tras instalar y cargar nuestra libreria correctamente, el siguiente paso es mostrar la matriz de graficas de correlación:

```r
correlations <- cor(dataCleaned[,1:65])
# create correlation plot
corrplot(correlations, method="circle")
```

Podemos ver que los sensores tienen una correlación directa con el sensor vecino y una correlación negativo cuando se trata de la medición de un sensor  en otro musculo. Para averiguar la mejor selección de variables a la hora de predecir la clase V65, ejecutamos el algoritmo **randomForest**

Como siempre, antes de ello, cargamos las librerias necesarias:

```r
if (!requireNamespace("caret"))
  install.packages("caret")
library(caret)
if (!requireNamespace("randomForest"))
  install.packages("randomForest")
library(randomForest)
if (!requireNamespace("doParallel"))
  install.packages("doParallel")
library(doParallel) #Paralelismo... # Probamos

parallel::detectCores()

cl <- makePSOCKcluster(2) # Probar en Colab

registerDoParallel(cl) 
```

Posteriormente ponemos a ejecutar el algoritmo:

```r
if (!requireNamespace("klaR"))
install.packages("klaR")

# treebagfuncs
set.seed(230520202)

options(warn=-1)

subsets <- c(1:64)

ctrl <- rfeControl(functions = treebagFuncs,
                   method = "repeatedcv",
                   repeats = 1, # Las primeras veces es recomendable hacer el problema lo más pequeño para reducir tiempos
                   verbose = FALSE)

lmProfile <- rfe(x=dataCleaned[, 1:64], y=dataCleaned[, 65],
                 sizes = subsets,
                 rfeControl = ctrl)

lmProfile
```

```r
Importancia <- varImp(lmProfile)
Importancia
```
